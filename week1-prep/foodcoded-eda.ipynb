```python
import pandas as pd
import numpy as np
np.set_printoptions(precision=2, suppress=True)
pd.options.display.float_format = '${:,.2f}'.format
```

Read in data in csv format as a Pandas data frame. A __DataFrame__ is a 2-dimensional labeled data structure with columns of potentially different types. 

```python
df = pd.read_csv('foodcoded.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['GPA'].value_counts()
```

```python
df['GPA'] = df['GPA'].str.replace(r'[^0-9.]', '')
df['GPA'].value_counts()
```

```python
df.isna().sum()
```

replace the empty string to `nan` so they can be remove in the next step

```python
df['GPA'].replace('', np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['GPA'])
```

```python
df['GPA'] = df['GPA'].astype('float')
```

```python
df.shape
```

Select only the numeric columns 

```python
df_num = df.select_dtypes('number')
```

```python
df_num.columns
```

```python
corr = df_num.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```
